## Defanging an IP Address

```python
address = '255.156.145.147'
for i in address:
    if i == '.':
        output = address.replace('.','[.]')
print(output)   
```

## Find Numbers with Even Number of Digits

```python
output = 0
nums = [901,482,1441]
for i in range(len(nums)):
    count = 0
    while nums[i]>0:
        nums[i] = nums[i]//10
        count+=1
    if count%2==0: 
        output+=1
print(output)
```

## Number of Good Pairs

```python
count = 0
nums = [1,2,3,1,1,3]
for i in range(len(nums)-1):
    for j in range(1,len(nums)):
        if nums[i]==nums[j] and i<j:
            count+=1
print(count)
```

## How Many Numbers Are Smaller Than the Current Number

```python
a =[]
nums = [8,1,2,2,3]
count = 0
for i in range(len(nums)):
    for j in range(len(nums)):
        if nums[i]>nums[j]:
            count+=1
    a.append(count)
    count = 0
print(a)
```

## Subtract the Product and Sum of Digits of an Integer

```python
a = 1
b = 0
n= 234
while n>0:
    c = n%10
    a *= c
    b += c
    n = n//10
print(a-b) 
```

## XOR Operation in an Array

```python
res = 0
n = 5
start = 0
for i in range(n):
    res ^= start + 2*i
print(res)
```
